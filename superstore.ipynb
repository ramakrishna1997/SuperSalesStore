## For each entry in dataset ship date >= order date

```python
def comp(ship_date,order_date):
    ship_date=ship_date.values.reshape((-1,1))
    order_date=order_date.values.reshape((-1,1))
    result=np.any((ship_date >= order_date),axis=1)
    return result
```

```python
sales["Ship_after_order"]=comp(sales["Ship_Date"],sales["Order_Date"])
```
